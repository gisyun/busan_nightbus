```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from glob import glob
import folium
```

```python
#glob()로 부산 중심관광지 데이터 불러오기
file_names = glob(".//중심관광지/*.csv")
busan_poi = pd.DataFrame() 


for file_name in file_names:
    poi = pd.read_csv(file_name, encoding = "EUC-KR")

    busan_poi = pd.concat([busan_poi, poi])


busan_poi
```

```python
#중심카테고리 분류 체계 확인

print(busan_poi['중심카테고리 명_대'].unique())
print(busan_poi['중심카테고리 명_중'].unique())
```

```python
#태블로 시각화 위해 합친 데이터셋 반출 

busan_poi.to_csv("./중심관광지/central_poi(total).csv")
```

```python
poi_coord = busan_poi[['중심 POI X 좌표', '중심 POI Y 좌표','중심관광지명']]
poi_coord
```

```python
poi_coord.columns = ['longitude', 'latitude']
poi_coord
```

```python
poi_coord.describe()
```

## Visualization

```python
poi_locations = poi_coord[['latitude', 'longitude']]
poi_coord_list = poi_locations.values.tolist()
poi_coord_list
```

```python
#poi 맵핑

map = folium.Map(location=[35.155063,129.063081], zoom_start=11)
for coord in poi_coord_list:
    folium.CircleMarker(coord,
                        radius=8,
                        color='b',
                        fill_color="blue",
                        popup="CircleMarker popup",
                        tooltip="CircleMarker tooltip",).add_to(map)

map
```

```python
# 중심관광지 히트맵 시각화

from folium.plugins import HeatMap, HeatMapWithTime

#hot_spot 데이터 히트맵으로 시각화

map_hm = folium.Map(location=[35.155063,129.063081],
                    zoom_start = 12) 
poi_coord['latitude'] = poi_coord['latitude'].astype(float) #float형태 변환 
poi_coord['longitude'] = poi_coord['longitude'].astype(float)
poi_coord = poi_coord[['latitude', 'longitude']]

#지도에 추가
HeatMap(poi_coord,gradient={0.2: 'blue', 0.4: 'lime', 0.6: 'yellow', 0.8: 'orange', 1: 'red'}, 
                min_opacity=0.05, 
                max_opacity=0.9, 
                radius=25).add_to(map_hm)

#확인 
map_hm
```

히트맵 분석을 통해 육안으로 확인한 결과 약 7개의 클러스터가 형성될 것을 예상할 수 있음 

## 클러스터링

```python
from sklearn.cluster import DBSCAN

db_df = poi_coord[['중심 POI X 좌표', '중심 POI Y 좌표']]
db = DBSCAN(eps=0.005, min_samples=7, metric='euclidean').fit(db_df)
labels = db.labels_

poi_coord['cluster'] = labels

print(poi_coord['cluster'].unique())
poi_coord
```

```python
poi_coord.to_csv('./please_db.csv')
```

```python
#dbscan 결과값을 통해 불리언 생성, outlier = False로 변경

core_samples_mask = np.zeros_like(db.labels_, dtype = bool)
core_samples_mask[db.core_sample_indices_] = True
```

```python
# dbscan 성능평가

n_clusters_ = len(set(labels)) - ( 1 if -1 in labels else 0)
n_noise = list(labels).count(-1)
print('# of clusters->', n_clusters_)
print('noise count->', n_noise)

#하.. 처참
```
